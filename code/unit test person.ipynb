```python
# use unittest
import unittest
from person import Person  # Import the Person class from person.py

class TestPerson(unittest.TestCase):
    def test_is_newborn(self):
        newborn = Person(age=0, employed=False)
        self.assertTrue(newborn.is_newborn())
        self.assertFalse(newborn.is_young())
        self.assertFalse(newborn.is_workforce())
        self.assertFalse(newborn.is_senior())
        self.assertTrue(newborn.is_unemployed())
        self.assertFalse(newborn.is_employed())

    def test_is_young(self):
        young = Person(age=10, employed=False)
        self.assertFalse(young.is_newborn())
        self.assertTrue(young.is_young())
        self.assertFalse(young.is_workforce())
        self.assertFalse(young.is_senior())
        self.assertTrue(young.is_unemployed())
        self.assertFalse(young.is_employed())

    def test_is_workforce(self):
        workforce = Person(age=25, employed=False)
        self.assertFalse(workforce.is_newborn())
        self.assertFalse(workforce.is_young())
        self.assertTrue(workforce.is_workforce())
        self.assertFalse(workforce.is_senior())
        self.assertTrue(workforce.is_unemployed())
        self.assertFalse(workforce.is_employed())

    def test_is_senior(self):
        senior = Person(age=70, employed=False)
        self.assertFalse(senior.is_newborn())
        self.assertFalse(senior.is_young())
        self.assertFalse(senior.is_workforce())
        self.assertTrue(senior.is_senior())
        self.assertTrue(senior.is_unemployed())
        self.assertFalse(senior.is_employed())

    def test_is_unemployed(self):
        unemployed = Person(age=30, employed=False)
        self.assertFalse(unemployed.is_unemployed())
        employed = Person(age=40, employed=True)
        self.assertFalse(employed.is_unemployed())

    def test_is_employed(self):
        employed = Person(age=50, employed=True)
        self.assertTrue(employed.is_employed())
        unemployed = Person(age=60, employed=False)
        self.assertFalse(unemployed.is_employed())

if __name__ == '__main__':
    unittest.main()
```
