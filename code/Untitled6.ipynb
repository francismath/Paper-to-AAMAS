```python
import random
import math

class Person:
    def __init__(self, age):
        self.age = age

    def update_age(self):
        self.age += 1


class Population:
    def __init__(self, newborns, young, workforce, retirees):
        self.newborns = newborns
        self.young = young
        self.workforce = workforce
        self.retirees = retirees

    def transition(self):
        # Update age for each person in the population
        self.update_age()

        # Calculate the number of babies born based on the sum of workforce and retirees
        total_population = self.workforce + self.retirees
        babies = math.ceil(len(total_population) * random.uniform(0.01, 0.03))  # Assume 1-3% of the total population are babies

        # Distribute babies to the newborns group
        len(self.newborns) += babies

        # Move individuals from one age group to another
        self.young += self.newborns
        self.newborns = 0

        self.workforce += self.young
        self.young = 0

        self.retirees += self.workforce
        self.workforce = 0

    def update_age(self):
        # Update age for each person in the population
        self.newborns = [Person(p.age + 1) for p in self.newborns]
        self.young = [Person(p.age + 1) for p in self.young]
        self.workforce = [Person(p.age + 1) for p in self.workforce]
        self.retirees = [Person(p.age + 1) for p in self.retirees]

    def print_population(self):
        print("Newborns: ", len(self.newborns))
        print("Young (1-14): ", len(self.young))
        print("Workforce (15-64): ", len(self.workforce))
        print("Retirees (65 and over): ", len(self.retirees))


# Example usage
newborns = [Person(random.randint(0, 1)) for _ in range(50)]  # Randomly assign age 0 or 1 to newborns
young = [Person(random.randint(1, 14)) for _ in range(500)]
workforce = [Person(random.randint(15, 64)) for _ in range(1000)]
retirees = [Person(random.randint(65, 80)) for _ in range(50)]

population = Population(newborns, young, workforce, retirees)
print("Initial population:")
population.print_population()

# Simulate population transition for a year
population.transition()
print("\nPopulation after one year:")
population.print_population()
```
