```python
# step 3: initialize the cohort of newborn 14:33 0207 
def count_age(age, group):
    num = 0
    for person in group:
        if person.age == age:
            num += 1
            
    return num

import random
import math
import itertools # join multiple lists
import matplotlib.pyplot as plt

# Parameters
ini_population = 1000
simulation_years = 50

birth_rate = 0.011 # assume only population 15 years and over reproduce 

death_rate = 0.01 # assume death only happens after senior with count_death = count_senior * death_rate
death_age = 80 #a simple alternative way for the senior is assume they die after 80 years old
retirement_age = 65

ini_workforce_rate = 0.7 # workforce rate = count 15-64/population

ini_aging_rate = 0.1 # aging_rate = count 65 and over/ workforce
ini_unemployment_rate = 0.05

class Person:
    def __init__(self, age, employed):
        self.age = age
        self.employed = employed
    
    def update_age(self):
        self.age += 1
        
    def is_newborn(self):
        return self.age == 0
    
    def is_young(self):
        return self.age < 15
    
    def is_workforce(self):
        return 14 < self.age < 65
    
    def is_senior(self):
        return self.age > 64
    
ini_workforce = ini_population * ini_workforce_rate # 700
ini_senior = ini_workforce * ini_aging_rate # 70
ini_unemp = ini_workforce * ini_unemployment_rate # 700*0.05=35
ini_young = ini_population - ini_workforce - ini_senior

population_coh = []

newborn_coh = []
young_coh = []
workforce_coh = []
senior_coh = []

unemp_coh = []
emp_coh = []

unem_rate = []

# iniize population
random.seed(1)
for i in range(ini_population):
    if i < ini_young:
        person = Person(age=random.randint(0,14), employed=False)
        young_coh.append(person)
    
    elif i < ini_young + ini_unemp:
        person = Person(age=random.randint(15,64), employed=False)
        unemp_coh.append(person)
        
    elif i < ini_young + ini_workforce:
        person = Person(age=random.randint(15,64), employed=True)
        emp_coh.append(person)
        
    else:
        person = Person(age=random.randint(65,80), employed=False)
        senior_coh.append(person)

'''
ini_num_newborn = math.ceil((ini_workforce + ini_senior)*birth_rate)
for i in range(ini_num_newborn):
    person = Person(age=0, employed=False)
    newborn_coh.append(person)

'''
workforce_coh = list(itertools.chain(unemp_coh, emp_coh))
cohorts = [[len(newborn_coh)], [len(young_coh)], [len(workforce_coh)], [len(senior_coh)]]
population_coh = list(itertools.chain(newborn_coh, young_coh, workforce_coh, senior_coh))

print(cohorts)

# Simulation loop
random.seed(2)
n = simulation_years - 45
for year in range(n):
    young_this_year = 0
    workforce_this_year = 0
    senior_this_year = 0
    newborn_this_year = 0

    print('count_14:', count_age(14, young_coh))
    print('count_64:', count_age(64, workforce_coh))
    
    for person in population_coh:      
                
        if person.is_newborn():
            newborn_this_year += 1
            
        if person.is_young():
            young_this_year += 1
            
        if person.is_workforce():
            workforce_this_year += 1
                    
        if person.is_senior():
            senior_this_year += 1
        
        person.update_age()
        
    
    newborn_this_year = math.ceil((workforce_this_year + senior_this_year)*birth_rate)

    for i in range(newborn_this_year):
        person = Person(age=0, employed=False)
        newborn_coh.append(person)
        
    for i in range(young_this_year):
        person = Person(age=random.randint(1,14), employed=False)
        young_coh.append(person)
        
    young_coh = list(itertools.chain(newborn_coh, young_coh))
    
    #newborn_this_year = len(newborn_coh)
    young_this_year = len(young_coh)
    print('new born at year', year, 'is:', newborn_this_year)
    print('workforce at year', year, 'is:', workforce_this_year)
    print('senior at year:', year, 'is:', senior_this_year)
    print('kids at year:', year, 'is:', young_this_year, '\n')
```
