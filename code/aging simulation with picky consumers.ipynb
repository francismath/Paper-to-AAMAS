```python
import random

class Firm:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def update_price(self):
        self.price += random.uniform(-0.5, 0.5)

    def sell_product(self):
        return self.price

class Consumer:
    def __init__(self, name, income, pickiness):
        self.name = name
        self.income = income
        self.pickiness = pickiness

    def buy_product(self, firm):
        if firm.price <= self.income and random.random() > self.pickiness:
            self.income -= firm.price
            print(f"{self.name} bought from {firm.name} at price {firm.price}")
            return True
        else:
            print(f"{self.name} did not buy from {firm.name}")
            return False

class Retiree(Consumer):
    def __init__(self, name, income, pickiness, buying_time):
        super().__init__(name, income, pickiness)
        self.buying_time = buying_time

    def buy_product(self, firm):
        if self.buying_time > 0:
            self.buying_time -= 1
            print(f"{self.name} is taking more time to buy from {firm.name}")
            return False
        else:
            return super().buy_product(firm)

def main():
    firms = [Firm("Firm 1", 10.0), Firm("Firm 2", 8.0), Firm("Firm 3", 12.0)]
    consumers = [Consumer("Consumer 1", 15.0, 0.5), Consumer("Consumer 2", 20.0, 0.7)]
    retirees = [Retiree("Retiree 1", 12.0, 0.8, 2), Retiree("Retiree 2", 10.0, 0.6, 3)]

    for _ in range(5):
        random.shuffle(firms)  # Randomize the order of firms

        for firm in firms:
            firm.update_price()

            for consumer in consumers:
                if consumer.buy_product(firm):
                    break

            for retiree in retirees:
                if retiree.buy_product(firm):
                    break

if __name__ == "__main__":
    main()
```
